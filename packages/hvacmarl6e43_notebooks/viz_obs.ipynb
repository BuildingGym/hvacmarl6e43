```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams.update({
    'backend': 'svg',
    'pgf.texsystem': 'lualatex',
    'pgf.rcfonts': False,
    'text.usetex': True,
    'font.family': 'serif',
    'font.serif': [],
})

import matplotlib.axis as maxis
import matplotlib.dates as mdates

def mpl_autofmt_datetime(axis: maxis.Axis):
    axis.set_major_locator(mdates.AutoDateLocator())
    axis.set_major_formatter(mdates.ConciseDateFormatter(axis.get_major_locator()))

from hvplot import hvPlot as Plot
import holoviews as hv
#hv.extension('bokeh', 'plotly', 'matplotlib')
hv.extension('matplotlib', 'bokeh')

import pandas as pd
```

```python
from pathlib import Path
from hvacmarl6e43_notebooks import notebook_storage_base

figurestorepath = Path(notebook_storage_base)
figurestorepath.mkdir(exist_ok=True)
```

```python
from hvacmarl6e43.experiments import baseline_experiments, rl_experiments

datas = {
    name: experiment.get_results().observations
    for name, experiment in {
        **baseline_experiments,
        **rl_experiments,
    }.items()
}
```

```python
datas_avg = dict()

for name, data in datas.items():
    df = pd.concat(data, axis='columns')
    df_avg = df.groupby(level=1, axis='columns').mean()
    datas_avg[name] = df_avg


datas_avg_elec_sorted = {
    key: df.sort_values(by='elec')
    for key, df in datas_avg.items()
}

hv.Overlay([
    Plot(df[df['occupancy'] != 0])
    .line(x='elec', y='pmv_abs', label=str(name))
    .opts(alpha=.8)
    for name, df in datas_avg_elec_sorted.items()
]) #.opts(width=1_000, legend_position='right')
```

```python
datas_avg_elec_sorted_rolling: dict[object, pd.DataFrame] = dict()
for key, df in datas_avg_elec_sorted.items():
    df = df[df['occupancy'] != 0]
    pmv_abs_rolling = df['pmv_abs'].rolling(window=20, min_periods=1)
    datas_avg_elec_sorted_rolling[key] = pd.DataFrame({
        'elec': df['elec'],
        'pmv_abs_rolling_mean': pmv_abs_rolling.mean(),
        'pmv_abs_rolling_min': pmv_abs_rolling.min(),
        'pmv_abs_rolling_max': pmv_abs_rolling.max(),
    })

fig = hv.Overlay([
    hv.Overlay([
        hv.Curve((df['elec'], df['pmv_abs_rolling_mean']), label=str(name)),
        hv.Area(
            (df['elec'], df['pmv_abs_rolling_min'], df['pmv_abs_rolling_max']), 
            vdims=['y_min', 'y_max'], 
            label=str(name),
        )
        .opts(xlabel=r'Power', ylabel='Absolute PMV (Rolling Avg.)')
        .opts(alpha=.1, line_width=0, linewidth=0)
    ])
    for name, df in datas_avg_elec_sorted_rolling.items()
]).opts() #.opts(width=1_000, legend_position='right')
hv.save(fig, figurestorepath / 'figure:elec-v-pmv-abs-rolling.pgf')
fig
```

```python
df_metrics_avg = pd.DataFrame([
    {
        'name': str(name),
        'elec_mean': df[df['occupancy'] != 0]['elec'].mean(),
        'pmv_abs_mean': df[df['occupancy'] != 0]['pmv_abs'].mean(),
    }
    for name, df in datas_avg.items()
])
```

```python
df = df_metrics_avg.sort_values(by=['elec_mean'])
fig = hv.Overlay([
    Plot(df).scatter(x='elec_mean', y='pmv_abs_mean', markers=True),
    hv.Labels(df, ['elec_mean', 'pmv_abs_mean'], 'name'),
])
hv.save(fig, figurestorepath / 'figure:elec-mean-v-pmv-abs-mean.pgf')
fig
```

```python
fig, (ax_elec, ax_pmv_abs_mean) = plt.subplots(
    nrows=1, ncols=2, 
    sharey=True,
    figsize=(6, 2),
)

bars_elec = ax_elec.barh(
    df_metrics_avg['name'], 
    df_metrics_avg['elec_mean'],
    color='salmon',
)
ax_elec.bar_label(
    bars_elec, 
    label_type='center',
)
ax_elec.set(xlabel='Power (Avg.)')

bars_pmv_abs_mean = ax_pmv_abs_mean.barh(
    df_metrics_avg['name'], 
    df_metrics_avg['pmv_abs_mean'],
    color='skyblue',
)
ax_pmv_abs_mean.bar_label(
    bars_pmv_abs_mean, 
    label_type='center',
)
ax_pmv_abs_mean.set(xlabel='PMV (Absolute Avg.)')

fig.tight_layout()
fig.savefig(figurestorepath / 'figure:avg-elec-pmv-abs-mean.pgf')
fig
```

```python
from collections import defaultdict

datas_eachroom = defaultdict(dict)
for name, data in datas.items():
    for room_id, df in data.items():
        df = pd.DataFrame(df)
        df['is_comfortable'] = df['pmv_abs'] <= .5
        datas_eachroom[room_id][name] = df
```

```python
from itertools import zip_longest

import matplotlib.pyplot as plt
from matplotlib.patches import Patch

start_date = pd.to_datetime('2002-07-08').tz_localize('UTC')
end_date = pd.to_datetime('2002-07-15').tz_localize('UTC')

def _mpl_shared_legend_patches(axes):
    return [
        Patch(color=color, label=label) 
        for (label, color) in dict.fromkeys([
            (label, handle.get_color())
            for ax in axes
            for handle, label in zip(*ax.get_legend_handles_labels())
        ])
    ]


n_cols = 2
n_rows = round(len(datas_eachroom) / n_cols)
fig = plt.figure(figsize=(8, 8))
subfig_data, subfig_legend = fig.subfigures(
    nrows=2, ncols=1,
    height_ratios=(.9, .1),
)

axes = subfig_data.subplots(
    n_rows, 
    n_cols, 
    sharex=True, sharey=True, 
)
for ax, data in zip_longest(
    axes.flat, 
    datas_eachroom.items(),
    fillvalue=None,
):
    if data is None:
        ax.axis('off')
        continue
    
    room_id, room_datas = data
    for experiment_name, df in room_datas.items():
        sorted_df = df.sort_index().loc[(df.index >= start_date) & (df.index <= end_date)]
        ax.plot(sorted_df.index, sorted_df['pmv_abs'], label=str(experiment_name))
    
    ax.set_title(str(room_id), loc='left')

    mpl_autofmt_datetime(ax.xaxis)

fig.supxlabel('Time')
fig.supylabel('PMV (Absolute)')
subfig_legend.legend(
    handles=(handles := _mpl_shared_legend_patches(axes.flat)), 
    ncol=len(handles),
    loc='center',
)
fig.tight_layout()

fig.savefig(figurestorepath / 'figure:time-v-pmv-abs.pgf')
fig
```

```python
comfort_datas = dict()
for name, data in datas.items():
    df = pd.DataFrame({
        room_id: 
            (df['pmv_abs'] <= .5)
            .resample('1h').mean()
        for room_id, df in data.items()
    })
    df_sum = df.mean(axis='columns')
    # TODO
    comfort_datas[name] = df_sum.iloc[:24]
```

```python
import matplotlib.ticker as mtick

fig, ax = plt.subplots(figsize=(8, 4))
ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.))
df = pd.DataFrame(comfort_datas)
df = df[(df != 0).any(axis='columns')]
date_offset = df.index[0]
df.index = (df.index - date_offset) / pd.Timedelta(hours=1)
df.plot.bar(
    ax=ax, 
    xlabel=f'Hours Since {date_offset}',
    ylabel='Proportion of Comfortable Timesteps',
)
fig.autofmt_xdate()
fig.tight_layout()
fig.savefig(figurestorepath / 'figure:time-v-comfort.pgf')
fig
```
